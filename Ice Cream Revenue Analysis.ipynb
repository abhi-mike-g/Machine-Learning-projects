```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
icecream_df = pd.read_csv('IceCreamData.csv')
```

```python
plt.figure(figsize = (13, 7))
sns.scatterplot(x = 'Temperature', y = 'Revenue', data = icecream_df)
plt.grid()
```

```python
plt.figure(figsize = (13, 7))
sns.regplot(x = 'Temperature', y = 'Revenue', data = icecream_df)
plt.grid()
```

```python
icecream_df
```

```python
X = icecream_df['Temperature']
y = icecream_df['Revenue']
```

```python
X = np.array(X)
y = np.array(y)
```

```python
X = X.reshape(-1, 1)
y = y.reshape(-1, 1)
```

```python
print(X.shape)

print(y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
Simplinreg = LinearRegression(fit_intercept = True)
Simplinreg.fit(X_train, y_train)
```

```python
print('linear model slope (m) = ', Simplinreg.coef_)
print('linear model Intercept (m) = ', Simplinreg.intercept_)
```

```python
plt.scatter(X_train, y_train, color = 'blue')
plt.plot(X_train, Simplinreg.predict(X_train), color = 'black')
plt.ylabel('Revenue[$]')
plt.xlabel('T[degC]')
plt.title('Revenue generated vs Temperature @Ice Cream')
plt.grid()
```

```python
accuracy_LinReg = Simplinreg.score(X_test, y_test)
accuracy_LinReg
```

```python
plt.scatter(X_test, y_test, color = 'green')
plt.plot(X_test, Simplinreg.predict(X_test), color = 'red')
plt.ylabel('Revenue[$]')
plt.xlabel('T[degC]')
plt.title('Revenue generated vs Temperature @Ice Cream')
plt.grid()
```
